```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
df_list = []
years = ['2010-11', '2012-13', '2013-14', '2014-15', '2015-16', '2016-17', '2017-18', '2018-19']
months = ['October', 'November', 'December', 'January', 'February', 'March', 'April']
columns = ['Starters','MP','FG','FGA','FG%','3P','3PA','3P%','FT','FTA','FT%','ORB','DRB','TRB','AST','STL','BLK','TOV','PF','PTS','+/-']

for h in range(0, len(years)):
    driver = webdriver.Chrome(executable_path='C:/Users/yesla/bin/chromedriver.exe')
    driver.get("https://www.basketball-reference.com/leagues/")
    driver.find_element_by_link_text(years[h]).click()
    driver.find_element_by_link_text("Schedule and Results").click()
    for i in range(0, len(months)):
        # Get all of the box scores for October in the 2010-2011 season
        driver.find_element_by_link_text(months[i]).click()
        allgames = driver.find_element_by_id("schedule")
        tr_list = allgames.find_elements_by_tag_name('tr')
    
        # First for loop to get the first 30 boxscores
        if months[i] != 'April':
            for j in range(1,len(tr_list)):
            # Use try statement to get passed the weird table formatting
                try:
                    # Click on the boxscores for the NBA games
                    allgames = driver.find_element_by_id("schedule")
                    tr_list = allgames.find_elements_by_tag_name('tr')
                    tr_list[j].find_element_by_link_text("Box Score").click()
    
                    # Get team names for later use
                    soup = BeautifulSoup(driver.page_source, 'html.parser')
                    table = soup.find(name='table', attrs={'id':'line_score'})
                    html_str = str(table)
                    df = pd.read_html(html_str)[0]
                    team1 = df['Unnamed: 0'][1]
                    team2 = df['Unnamed: 0'][2]
    
                    # Extract and transform first teams boxscore
                    table1 = soup.find(name='table', attrs={'id':'box-'+team1+'-game-basic'})
                    html_str1 = str(table1)
                    df1 = pd.read_html(html_str1)[0]
                    df1 = pd.DataFrame(df1.to_records())
                    first_column = df1.columns[0]
                    df1 = df1.drop([first_column], axis=1)
                    df1.columns = columns
                    df1 = df1[(df1['Starters'] != 'Reserves') & (df1['Starters'] != 'Team Totals') & (df1['MP'] != 'Did Not Play')]
    
                    # Extract and transform second teams boxscore
                    table2 = soup.find(name='table', attrs={'id':'box-'+team2+'-game-basic'})
                    html_str2 = str(table2)
                    df2 = pd.read_html(html_str2)[0]
                    df2 = pd.DataFrame(df2.to_records())
                    first_column = df2.columns[0]
                    df2 = df2.drop([first_column], axis=1)
                    df2.columns = columns
                    df2 = df2[(df2['Starters'] != 'Reserves') & (df2['Starters'] != 'Team Totals') & (df2['MP'] != 'Did Not Play')]
    
                    # Concatenate both boxscores then append to empty list
                    df_comp = pd.concat([df1,df2])
                    df_list.append(df_comp)
                    driver.back()
                except:
                    pass
        else:
            for j in range(60,len(tr_list)):
            # Use try statement to get passed the weird table formatting
                try:
                    # Click on the boxscores for the NBA games
                    allgames = driver.find_element_by_id("schedule")
                    tr_list = allgames.find_elements_by_tag_name('tr')
                    tr_list[j].find_element_by_link_text("Box Score").click()
    
                    # Get team names for later use
                    soup = BeautifulSoup(driver.page_source, 'html.parser')
                    table = soup.find(name='table', attrs={'id':'line_score'})
                    html_str = str(table)
                    df = pd.read_html(html_str)[0]
                    team1 = df['Unnamed: 0'][1]
                    team2 = df['Unnamed: 0'][2]
    
                    # Extract and transform first teams boxscore
                    table1 = soup.find(name='table', attrs={'id':'box-'+team1+'-game-basic'})
                    html_str1 = str(table1)
                    df1 = pd.read_html(html_str1)[0]
                    df1 = pd.DataFrame(df1.to_records())
                    first_column = df1.columns[0]
                    df1 = df1.drop([first_column], axis=1)
                    df1.columns = columns
                    df1 = df1[(df1['Starters'] != 'Reserves') & (df1['Starters'] != 'Team Totals') & (df1['MP'] != 'Did Not Play')]
    
                    # Extract and transform second teams boxscore
                    table2 = soup.find(name='table', attrs={'id':'box-'+team2+'-game-basic'})
                    html_str2 = str(table2)
                    df2 = pd.read_html(html_str2)[0]
                    df2 = pd.DataFrame(df2.to_records())
                    first_column = df2.columns[0]
                    df2 = df2.drop([first_column], axis=1)
                    df2.columns = columns
                    df2 = df2[(df2['Starters'] != 'Reserves') & (df2['Starters'] != 'Team Totals') & (df2['MP'] != 'Did Not Play')]
    
                    # Concatenate both boxscores then append to empty list
                    df_comp = pd.concat([df1,df2])
                    df_list.append(df_comp)
                    driver.back()
                except:
                    break           
```
