# Data Cleaning & Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
# Read in 2011-2012 NBA regular season box score data
df = pd.read_csv('df-2011-12.csv')
df.head(100)
```

```python
# Info of the data frame gives me some ideas on how to clean up the data frame
df.info()
```

```python
# Filter out all of the NaN values in the 'FG' column
df = df[df['FG'].notnull() == True]
```

```python
# Change FT%, 3P%, and FG% NaN values into 0's
df.loc[df['FT%'].isnull(), 'FT%'] = 0
df.loc[df['3P%'].isnull(), '3P%'] = 0
df.loc[df['FG%'].isnull(), 'FG%'] = 0
df.head(5)
```

```python
# Change the MP into a number by changing the ':' into a '.' and casting the column as a number
df['MP'] = pd.to_numeric(df['MP'].str.replace(':','.'))
df.head()
```

```python
# Look at the spread of the target varaible 'PTS'
sns.distplot(df['PTS'])
```

```python
# Correlation heat map to look over the different columns
corr = df.corr()

sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns)
```

```python
# Jointplot that shows the relationship between  'MP' and 'PTS' using hexagonal bins
sns.jointplot(df['MP'], df['PTS'], kind = 'hex')
```

```python
# Pairplot of all of the columns that will be used in the modeling before normalizing
df_minus = df.drop(['Starters', '+/-'], axis = 1)
sns.pairplot(df_minus)
```

```python
#Remove a couple more columns that are directly correlated to points in a game
df_cleaned = df.drop(['Starters', '+/-', 'FG', 'FGA', '3P', '3PA', 'FT', 'FTA'], axis = 1)

# Keep to unnomalize later
max_pts = df_cleaned['PTS'].max()
min_pts = df_cleaned['PTS'].min()

#Normalize 
df_mp = (df_cleaned['MP'] - df_cleaned['MP'].min()) / (df_cleaned['MP'].max() - df_cleaned['MP'].min())
df_fgp = (df_cleaned['FG%'] - df_cleaned['FG%'].min()) / (df_cleaned['FG%'].max() - df_cleaned['FG%'].min())
df_tpp = (df_cleaned['3P%'] - df_cleaned['3P%'].min()) / (df_cleaned['3P%'].max() - df_cleaned['3P%'].min())
df_ftp = (df_cleaned['FT%'] - df_cleaned['FT%'].min()) / (df_cleaned['FT%'].max() - df_cleaned['FT%'].min())
df_trb = (df_cleaned['TRB'] - df_cleaned['TRB'].min()) / (df_cleaned['TRB'].max() - df_cleaned['TRB'].min())
df_ast = (df_cleaned['AST'] - df_cleaned['AST'].min()) / (df_cleaned['AST'].max() - df_cleaned['AST'].min())
df_stl = (df_cleaned['STL'] - df_cleaned['STL'].min()) / (df_cleaned['STL'].max() - df_cleaned['STL'].min())
df_blk = (df_cleaned['BLK'] - df_cleaned['BLK'].min()) / (df_cleaned['BLK'].max() - df_cleaned['BLK'].min())
df_tov = (df_cleaned['TOV'] - df_cleaned['TOV'].min()) / (df_cleaned['TOV'].max() - df_cleaned['TOV'].min())
df_pf = (df_cleaned['PF'] - df_cleaned['PF'].min()) / (df_cleaned['PF'].max() - df_cleaned['PF'].min())
df_pts = (df_cleaned['PTS'] - df_cleaned['PTS'].min()) / (df_cleaned['PTS'].max() - df_cleaned['PTS'].min())

df_norm = pd.DataFrame({'PTS':df_pts, 'MP':df_mp, 'FG%':df_fgp, '3P%':df_tpp, 'FT%':df_ftp, 'TRB':df_trb, 'AST':df_ast,
                        'STL':df_stl, 'BLK':df_blk, 'TOV':df_tov, 'PF':df_pf, 'PTS':df_pts})
```
